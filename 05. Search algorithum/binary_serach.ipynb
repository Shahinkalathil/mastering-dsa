```python
# Binary Search Syntax
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

arr = [1, 2, 3, 6,5, 6, 7, 8, 9]
print(arr)
target = 
result = binary_search(arr, target)
if result != -1:    
  print(f"Element found at index: {result}")
else:    
  print("Element not found in the array.")           
```

```python
""" 
Question 1: Find Element Index
Find index of target in sorted array

[1 2 3 4 5], target = 3
mid = 3 → go left
mid = 2 → FOUND
"""
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

arr = [1, 2, 3, 4, 5]
print(arr)
target = 3
result = binary_search(arr, target)
if result != -1:    
  print(f"Element found at index: {result}")
else:    
  print("Element not found in the array.")           
```

```python
""" 
Question 2: Find First Occurrence
Find first position of target (duplicates exist)

[1 2 3 4 5], target = 3
mid = 3 → go left
mid = 2 → FOUND
"""
def first_occurrence(arr, target):
    left, right = 0, len(arr)-1
    result = -1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            result = mid
            right = mid - 1  
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return result


print(first_occurrence([1,2,2,2,3], 2))
```

```python
""" 
Question 3: Find Square Root (Integer)
Find floor of √n

n = 16 → 4
n = 10 → 3
"""
def sqrt_binary(n):
    left, right = 0, n
    ans = 0

    while left <= right:
        mid = (left + right) // 2

        if mid * mid <= n:
            ans = mid
            left = mid + 1
        else:
            right = mid - 1

    return ans


print(sqrt_binary(10)) 
```
