1. Linear Search Syntax
2. Find index of target
3. Return True/False
4. Find largest element using linear scan

```python
""" 
Linear Search Syntax
Works on any array (sorted or unsorted)
Find index of target

"""

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

arr = [4, 2, 1, 7, 9]
target = 7

print(linear_search(arr, target))
```

```python
""" 
Question 2: Check Element Exists
Return True/False
[3 5 8 2], target = 6

3 ❌  
5 ❌  
8 ❌  
2 ❌  
→ Not found
"""
def is_present(arr, target):
    for num in arr:
        if num == target:
            return True
    return False

print(is_present([3,5,8,2], 6))  
```

```python
""" 
Question 3: Find Maximum Element
Find largest element using linear scan
[4 2 9 1]

Start max = 4  
Compare 2 → no  
Compare 9 → update  
Compare 1 → no  

Result = 9
"""
def is_present(arr, target):
    for num in arr:
        if num == target:
            return True
    return False

print(is_present([3,5,8,2], 6))
```
