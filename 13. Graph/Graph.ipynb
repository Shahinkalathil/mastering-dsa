# fundamentals of the Graph Data Structure
1. Graph Representation (Adjacency List)
2. Degree of Each Node
3. Count Number of Edges

```python
# Graph Representation (Adjacency List)
graph = {
    1: [2, 3],
    2: [1, 4],
    3: [1],
    4: [2]
}

print("Graph Representation:")
for node in graph:
    print(node, "->", graph[node])
```

```python
# Degree of Each Node
def node_degrees(graph):
    print("Node Degrees:")
    for node in graph:
        print(f"Node {node} -> Degree {len(graph[node])}")

graph = {
    1: [2, 3],
    2: [1, 4],
    3: [1],
    4: [2]
}

node_degrees(graph)
```

```python
# Count Number of Edges
def count_edges(graph):
    total = 0
    for node in graph:
        total += len(graph[node])
    return total // 2   # undirected graph

graph = {
    1: [2, 3],
    2: [1, 4],
    3: [1],
    4: [2]
}

print("Total Edges:", count_edges(graph))
```
