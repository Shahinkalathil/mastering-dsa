# Queue using Linked List
1. Enqueue (Add element)
2. Dequeue (Remove element)
3. Display queue
4. Peek (Front element)
5. Check if empty

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data):
        new_node = Node(data)

        if self.rear is None:
            self.front = self.rear = new_node
            return 
        
        self.rear.next = new_node
        self.rear = new_node

    def dequeue(self):
        if self.front is None:
            return "Queue is Empty"

        temp = self.front
        self.front = temp.next

        if self.front is None:
            self.rear = None

        return temp.data

    def display(self):
        if self.front is None:
            print("Queue is Empty right now")
            return 
        temp = self.front
        while temp:
            print(temp.data, end="  ->  ")
            temp = temp.next
        print("None")

    def peek(self):
        if self.front is None:
            return "Queue is Empty right now"
        return self.front.data


q = Queue()
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
q.enqueue(40)
q.enqueue(50)
q.enqueue(60)
q.display()
print(q.dequeue())
q.display()
print(q.peek())
```
