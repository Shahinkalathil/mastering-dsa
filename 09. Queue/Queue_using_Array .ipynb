# Queue using Array (Python List)
1. Enqueue (Add element at rear)
2. Dequeue (Remove element from front)
3. Peek (Front element)
4. Check if empty
5. Display queue

```python
class Queue:
    def __init__(self):
        self.queue = []

    def enqueue(self, data):
        self.queue.append(data)
        print(f"{data} added to the queue")

    def dequeue(self):
        if self.empty():
            return f"The queue is empty"
        return self.queue.pop(0)
    
    def peek(self):
        if self.empty():
            return f"The queue is empty"
        return self.queue[0]
    
    def empty(self):
        return len(self.queue) == 0

    def display(self):
        if self.empty():
            return f"The queue is empty"
        else:
            return f"The array of queue: {self.queue}"

queue = Queue()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
queue.enqueue(40)
queue.enqueue(50)

queue.display()
print(queue.dequeue())
print(queue.peek())
print(queue.display())
```
