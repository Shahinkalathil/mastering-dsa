# Stack using Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None
    
    def travese(self):
        temp = self.top
        while temp:
            print(temp.data, end="  -->  ")
            temp = temp.next
        print("None")

    def PUSH(self, data):
        new_data = Node(data)
        new_data.next = self.top
        self.top = new_data

    def POP(self):
        if self.top is None:
            return "The Stack is Empty"
        removed = self.top.data
        self.top = self.top.next
        return f"the value {removed} is remove from the stack"

    def peek(self):
        if self.top is None:
            return "stack is Empty"
        return f"The Peek value in the linked list is {self.top.data}"

stack = Stack()
stack.PUSH(10)
stack.PUSH(20)
stack.PUSH(30)
stack.PUSH(40)  

stack.travese()
print(stack.POP())

stack.travese()
print(stack.peek())
stack.POP()
stack.travese()
stack.PUSH(50)
stack.travese() 
```
