# Stack (Python)
1. PUSH
2. POP
3. PEEK
4. EMPTY
5. SIZE

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, data):
        self.stack.append(data)
    
    def pop(self):
        if not self.empty():
            return self.stack.pop()
        return "Stack is Empty"
            
    def peek(self):
        if not self.empty():
            return self.stack[-1]
        return "Stack is Empty"

    def empty(self):
        return len(self.stack) == 0

    def size(self):
        return len(self.stack)


s = Stack()

s.push(10)
s.push(20)
s.push(30)

print("Stack:", s.stack)           # [10, 20, 30]
print("Stack size:", s.size())     # 3
print("Top element:", s.peek())    # 30
print("Removed:", s.pop())         # 30
```
