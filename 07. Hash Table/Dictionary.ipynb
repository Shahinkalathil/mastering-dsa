# Dictionary (Hash Table) Complexities
1. Create a Dictionary
2. Access Value
3. Add / Update Value
4. Delete Element
5. Check Key Exists
6. Loop Through Dictionary
7. Count Frequency
8. Merge Two Dictionaries
9. Find Key with Maximum Value
10. Remove Duplicates Using Dict

```python
print("1. Create a Dictionary")
student = {
    "name" : "shahin",
    "age" : 22,
}
print(student)
print()

print("2. Access Value")
print(student.get('name'))
print(student["age"])
print()

print("3. Add && Update Value")
student["cousre"] = "full stack develpemnt"
student["age"] = 22
print(student)
print()

print("4. Delete Element")
del student["cousre"]
# student.pop("age")
print(student)
print()

print("5. Check Key Exists")
if "name" in student:
    print("the student is exists")
else:
    print("not exists")
print()

print("6. Loop Through Dictionary - Traverse")
for key, value in student.items():
    print(f"stuents key: {key}, value: {value}")
print()

print("7. Count Frequency")
arr = [1, 2, 2, 3, 1, 1]
fre = {}

for num in arr:
    fre[num] = fre.get(num , 0) + 1

print(fre)
print()

print("8. Merge Two Dictionaries")
d1 = {"a": 1}
d2 = {"b": 2}

merge = {**d1, **d2}
print(merge)
print()

print("9. Find Key with Maximum Value")
max_value_key = max(merge, key=merge.get)
print(max_value_key)
print()

print("10. Remove Duplicates Using Dict")
arr = [1, 2, 2, 3, 1]
unique = list(dict.fromkeys(arr))
print(unique)
print()
```
