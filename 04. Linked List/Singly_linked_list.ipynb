```python
# 1. Basics of the linked list 
class Node:
    def __init__(self, data):
        self.data = data 
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None


first_node = Node("shahin")
second_node = Node("Luttu")
third_node = Node(30)
print(first_node.data, second_node.data , third_node.data)

# Connecting nodes
first_node.next = second_node
second_node.next = third_node

ll = LinkedList()
ll.head = first_node

current = ll.head

while current:
    print(current.data)
    current = current.next
```

```python
# 2. Print all elements of the linked list.(Traversal)
def traverse(self):
    temp = self.head
    while temp:
        print(temp.data, end="]----[")
        temp = temp.next
```

```python
# 3. Insert a new node at the start of the linked list.(Insert at Beginning)
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_begin(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def traverse(self):
        temp = self.head
        while temp:
            print(temp.data, end="]----[")
            temp = temp.next

insert_first = LinkedList()
insert_first.insert_begin(50)
insert_first.insert_begin(40)
insert_first.insert_begin(30)
insert_first.insert_begin(20)
insert_first.insert_begin(10)

insert_first.traverse()

```

```python
# 4. Insert a node at the end of the linked list.(Insert at End (Append))
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Linkedlist:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return 
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
    
    def traverse(self):
        temp = self.head
        while temp:
            print(temp.data)
            print(temp.next, "Address - next")
            print(temp, "Head")
            temp = temp.next


append_node = Linkedlist()
append_node.append(50)
append_node.append(40)
append_node.append(30)
append_node.append(20)
append_node.append(10)
append_node.traverse()
```

```python
# 5. Insert a node at a given position. (Insert at Specific Position)
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class linkedlist:
    def __init__(self):
        self.head = None

    def insert_position(self, pos, data):
        new_node = Node(data)
        # Insert at beginning
        if pos == 0:
            new_node.next = self.head
            self.head = new_node
            return
        
        temp = self.head
        for _ in range(pos - 1):
            if temp is None:
                print("Position out of range")
                return
            temp = temp.next

        new_node.next = temp.next
        temp.next = new_node

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return 
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def traverse(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp=temp.next
        
append_node = linkedlist()
append_node.append(50)
append_node.append(30)
append_node.append(20)

append_node.traverse()
append_node.insert_position(0,40)
print()
append_node.traverse()
```
