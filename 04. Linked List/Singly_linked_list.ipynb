```python
# 3. Insert a new node at the start of the linked list.(Insert at Beginning)
class Node:
	def __init__(self, data):
		self.data = data
		self.next = None

class LinkedList:
	def __init__(self):
		self.head = None

	def insert_begin(self, data):
		new_node = Node(data)
		new_node.next = self.head
		self.head = new_node

	def traverse(self):
		temp = self.head
		while temp:
			print(temp.data, end="]----[")
			temp = temp.next

insert_first = LinkedList()
insert_first.insert_begin(50)
insert_first.insert_begin(40)
insert_first.insert_begin(30)
insert_first.insert_begin(20)
insert_first.insert_begin(10)

insert_first.traverse()

```

```python
class Node:
	def __init__(self, data):
		self.data = data
		self.next = None


def array_to_linked_list(arr):
	if not arr:
		return None

	head = Node(arr[0])
	current = head

	for i in range(1, len(arr)):
		current.next = Node(arr[i])
		current = current.next

	return head

def print_list(head):
	temp = head
	while temp:
		print(temp.data, end=" → ")
		temp = temp.next
	print("None")

arr = [1, 2, 3, 4]
head = array_to_linked_list(arr)
print_list(head)
```

```python
# 🔹 Linked List Implementation with All Basic Operations

class Node:
	def __init__(self, data):
		self.data = data
		self.next = None


class Linked_list:
	def __init__(self):
		self.head = None

	# 1️⃣ Traverse (Print)
	def traverse(self):
		temp = self.head
		while temp:
			print(temp.data, end="]  [")
			temp = temp.next
		print("None")

	# 2️⃣ Insert at Beginning
	def insert_beginig(self, data):
		new_node = Node(data)
		if self.head:
			new_node.next = self.head
		self.head = new_node

	# 3️⃣ Insert at End
	def append(self, data):
		new_node = Node(data)
		if self.head is None:
			self.head = new_node
			return
		temp = self.head
		while temp.next:
			temp = temp.next
		temp.next = new_node

	# 4️⃣ Insert at Position (0-based index)
	def insert_pos(self, pos, data):
		new_node = Node(data)

		if pos == 0:
			self.insert_beginig(data)
			return

		temp = self.head
		for _ in range(pos - 1):
			if temp is None:
				print("Index out of range")
				return
			temp = temp.next

		print(temp.data, "temp data")
		print(temp.next.data, "temp next")

		if temp is None:
			print("Index out of range")
			return
		
		new_node.next = temp.next
		temp.next = new_node

	# 5️⃣ Delete from Beginning
	def delete_beigning(self):
		if self.head:
			self.head = self.head.next

	# 6️⃣ Delete from End
	def delect_end(self):
		if self.head is None:
			return

		if self.head.next is None:
			self.head = None
			return

		temp = self.head
		while temp.next.next:
			temp = temp.next

		temp.next = None

	# 🔟 Find Middle (Slow & Fast Pointer)
	def find_mid(self):
		fast = self.head
		slow = self.head

		while fast and fast.next:
			slow = slow.next
			fast = fast.next.next

		return slow.data if slow else None

	# 7️⃣ Search Element
	def search(self, key):
		temp = self.head
		i = 0

		while temp:
			if temp.data == key:
				return f"Value {key} found at index {i}"
			temp = temp.next
			i += 1

		return f"Value {key} not found"

	# 8️⃣ Find Length
	def lenght(self):
		count = 0
		temp = self.head

		while temp:
			count += 1
			temp = temp.next

		return count

	# 9️⃣ Reverse Linked List
	def reverse(self):
		prev = None
		curr = self.head

		while curr:
			nxt = curr.next
			curr.next = prev
			prev = curr
			curr = nxt

		self.head = prev


ll = Linked_list()

ll.append(10)
ll.append(20)
ll.append(30)
ll.append(40)
ll.append(50)

ll.traverse()

ll.insert_beginig(5)
ll.traverse()

ll.insert_pos(3, 25)
ll.traverse()

ll.delete_beigning()
ll.traverse()

ll.delect_end()
ll.traverse()

print("Middle:", ll.find_mid())
print(ll.search(40))
print("Length:", ll.lenght())

ll.reverse()
ll.traverse()
```
