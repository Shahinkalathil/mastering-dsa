#  Doubly Linked List Implementation with All Basic Operations
1. 1️⃣ Traverse (Forward)
2. 2️⃣ Insert at Beginning
3. 3️⃣ Insert at End
3. 4️⃣ Insert at Position (0-based index)

```python
#  Node Class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


#  Doubly Linked List
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # 1️⃣ Traverse (Forward)
    def traverse(self):
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next
        print("None")

    # 2️⃣ Insert at Beginning
    def insert_at_begin(self, data):
        new_node = Node(data)

        if self.head:
            self.head.prev = new_node
            new_node.next = self.head

        self.head = new_node

    # 3️⃣ Insert at End
    def insert_at_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node
        new_node.prev = temp

    # 4️⃣ Insert at Position (0-based index)
    def insert_at_pos(self, pos, data):
        new_node = Node(data)

        # Insert at beginning
        if pos == 0:
            self.insert_at_begin(data)
            return

        temp = self.head

        # Traverse to (pos-1)
        for _ in range(pos - 1):
            if temp is None:
                print("Position out of range")
                return
            temp = temp.next

        if temp is None:
            print("Position out of range")
            return

        # Insert node
        new_node.next = temp.next
        new_node.prev = temp

        if temp.next:
            temp.next.prev = new_node

        temp.next = new_node


# 🔹 TESTING
dll = DoublyLinkedList()

# Insert values: 10, 20, 40, 50, 60
dll.insert_at_end(10)
dll.insert_at_end(20)
dll.insert_at_end(40)
dll.insert_at_end(50)
dll.insert_at_end(60)

print("Initial list:")
dll.traverse()

# Insert 30 at position 2 (0-based index)
dll.insert_at_pos(2, 30)

print("After inserting 30 at position 2:")
dll.traverse()

# Insert at beginning
dll.insert_at_begin(5)
print("After inserting at beginning:")
dll.traverse()

# Insert at end
dll.insert_at_end(70)
print("After inserting at end:")
dll.traverse()
```
