#  Doubly Linked List Implementation with All Basic Operations

2. 2️⃣ Insert at Beginning
3. 3️⃣ Insert at End
4. 4️⃣ Insert at Position (0-based index)
5. 5️⃣ Delete from Beginning
6. 6️⃣ Delete from End
7. 7️⃣ Search Element
8. 8️⃣ Find Length
9. 9️⃣ Reverse Doubly Linked Lis
10. 🔟 Find Middle (Slow & Fast Pointer)

```python
# 🔹 Node Class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


# 🔹 Doubly Linked List
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # 1️⃣ Traverse (Forward)
    def traverse(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ]-> [ ")
            temp = temp.next
        print("None")

    # 2️⃣ Insert at Beginning
    def insert_at_begin(self, data):
        new_node = Node(data)

        if self.head:
            self.head.prev = new_node
            new_node.next = self.head

        self.head = new_node

    # 3️⃣ Insert at End
    def insert_at_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node
        new_node.prev = temp

    # 4️⃣ Insert at Position (0-based index)
    def insert_at_pos(self, pos, data):
        if pos == 0:
            self.insert_at_begin(data)
            return

        new_node = Node(data)
        temp = self.head

        for _ in range(pos - 1):
            if temp is None:
                print("Position out of range")
                return
            temp = temp.next

        if temp is None:
            print("Position out of range")
            return

        new_node.next = temp.next
        new_node.prev = temp

        if temp.next:
            temp.next.prev = new_node

        temp.next = new_node

    # 5️⃣ Delete from Beginning
    def delete_from_begin(self):
        if self.head is None:
            return

        if self.head.next:
            self.head = self.head.next
            self.head.prev = None   
        else:
            self.head = None

    # 6️⃣ Delete from End
    def delete_from_end(self):
        if self.head is None:
            return

        if self.head.next is None:
            self.head = None
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.prev.next = None

    # 7️⃣ Search Element
    def search(self, key):
        temp = self.head
        i = 0

        while temp:
            if temp.data == key:
                return f"Value found at index {i}: {temp.data}"
            temp = temp.next
            i += 1

        return "Value not found"

    # 8️⃣ Find Length
    def length_of_linkelist(self):
        count = 0
        temp = self.head

        while temp:
            count += 1
            temp = temp.next

        return count

    # 9️⃣ Reverse Doubly Linked List
    def reverse(self):
        temp = None
        curr = self.head

        while curr:
            temp = curr.prev
            curr.prev = curr.next
            curr.next = temp
            curr = curr.prev            

        if temp:
            self.head = temp.prev   
            
	# 🔟 Find Middle (Slow & Fast Pointer)
    def find_middle(self):
        fast = self.head
        slow = self.head
        
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow.data if slow else None
    



#  Testing
dll = DoublyLinkedList()

# Insert values
dll.insert_at_end(10)
dll.insert_at_end(20)
dll.insert_at_end(40)
dll.insert_at_end(50)
dll.insert_at_end(60)

print("Initial list:")
dll.traverse()
print()

# Insert at position
dll.insert_at_pos(2, 30)
print("After inserting 30 at position 2:")
dll.traverse()
print()

# Insert at beginning
dll.insert_at_begin(5)
print("After inserting at beginning:")
dll.traverse()
print()

# Insert at end
dll.insert_at_end(70)
print("After inserting at end:")
dll.traverse()
print()

# Delete from beginning
dll.delete_from_begin()
print("After delete_from_begin:")
dll.traverse()
print()

# Delete from end
dll.delete_from_end()
print("After delete_from_end:")
dll.traverse()
print()

# Search
print("Search Element:")
print(dll.search(30))
print()

# Length
print("Length of list:")
print(dll.length_of_linkelist())
print()

# Reverse
print("Reversed list:")
dll.reverse()
dll.traverse()
print()

print("find_middle")
print(dll.find_middle())
```
