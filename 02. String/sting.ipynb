```python
""" 
🧩 1. Replace Each Character by N-th Position
📝 Problem

Write a function to replace each alphabet with the character that appears n positions ahead.

👉 Wrap around if needed (z → a)
✅ Example
Input:  s = "abc", n = 2
Output: "cde"
"""
def shift_string(s, n):
    result = ""
    for char in s:
        if char.isalpha():
            base = 'a' if char.lower() else "A"
            new_char = chr(ord(char) - ord(base) + n % 26 + ord(base))
            result += new_char
        else:
            result += char
    return result
n = 3
s = "aAc 12B"
print(shift_string(s, n))

```

```python
""" 
🧩 2. Reverse Only Vowels in a String
📝 Problem

Reverse only the vowels in a string

✅ Example
Input:  "hello"
Output: "holle"
"""
def reverse_vowels(s):
    s = list(s)
    print(s)
    vowels = "aeiouAEIOU"
    
    left , right = 0, len(s)-1
    while left < right:
        if s[left] not in vowels:
            left += 1
        elif s[right] not in vowels:
            right -= 1
        else:
            s[left], s[right] = s[right], s[left]
            left += 1
            right -= 1
    return "".join(s)


word = 'my name is shahin'
print(word)
print(reverse_vowels(word))
```

```python
""" 
🧩 3. First Non-Repeating Character
📝 Problem

Find the first character that does not repeat

✅ Example
Input:  "aabbcde"
Output: "c"
"""

def first_unique(s):
    freq = {}
    
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    
    for char in s:
        if freq[char] == 1:
            return f"the fisrt unique letter is {char}"
    return f"There is no unique letter or stirng is empty"


word = "aabbcde"
print(first_unique(word))
```
