# 🔥 SIMPLE ARRAY INTERVIEW QUESTIONS (SINGLE FILE)
1. 1️⃣ Find Maximum Element
2. 2️⃣ Reverse Array (In-place)
3. 3️⃣ Check if Array is Sorted
4. 4️⃣ Find Second Largest Element
5. 5️⃣ Remove Duplicates (Sorted Array)

```python
# 1️⃣ Find Maximum Element

def max_value(arr):
    max_digit = arr[0]
    for val in arr:
        if val > max_digit:
            max_digit = val
    return max_digit


arr = [1,2,3,4,5]
print(max_value(arr))
```

```python
# 2️⃣ Reverse Array (In-place)
def reverse(arr):
    reverse = []
    for i in range(len(arr)-1, -1, -1):
        reverse.append(arr[i])
    return reverse

def reverse_pointer(arr):
    left , right = 0, len(arr)-1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr
         

arr = [1,2,3,4,5]
print(reverse(arr))
print(reverse_pointer(arr))
```

```python
# 4️⃣ Find Second Largest Element
def second_largest(arr):
    large = float("-inf")
    second = float('-inf')
 
    for i in arr:
        if i > large:
            second = large
            large = i
        elif i > second and i != second:
            second = i
    return second
arr = [1,2,3,4,5]
print(second_largest(arr))
```

```python
# 5️⃣ Remove Duplicates (Sorted Array)
def remove_duplicates(arr):
	if arr is None:
		return []
	
	result  = [arr[0]]
	for i in range(1, len(arr)):
		if arr[i] != arr[i-1]:
			result.append(arr[i])
		
	return result

sorted_arr = [1, 1, 2, 2, 3, 3]
print(remove_duplicates(sorted_arr))
```
