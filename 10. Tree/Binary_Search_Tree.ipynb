```python
"""
1. Full BST (Insert, Search, Delete, Traversals)
a. Insert

"""
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None
    
    def insert(self, data):
        self.root =  self._insert(self.root, data)

    def _insert(self, node, data):

        if node is None:
            return Node(data)
        
        if data < node.data:
            node.left = self._insert(node.left, data)
        elif data > node.data:
            node.right = self._insert(node.right, data)

        return node
    
    def inorder(self, node):
        if node is None:
            return 
        
        self.inorder(node.left)
        print(node.data, end=" ")
        self.inorder(node.right)
        
        
    
bst_value = [
    50,
    30, 70,
    20, 40, 60, 80,
    10, 25, 35, 45, 55, 65, 75, 90,
    5, 15, 22, 27, 33, 37, 43, 47,
    53, 57, 63, 67, 73, 77, 85, 95
]

tree = BST()
for val in bst_value:
    tree.insert(val)

tree.inorder(tree.root)
```

```python
""" 
b. Search (Recursive)
---------------------------
Time Complexity
Case	Time
Best	O(1) (value is at the root)
Average	O(log n)
Worst	O(n) (skewed BST)

Space:
Recursive → O(h)
Iterative → O(1)

"""
def search(self, data): # Recursive
    return self._search(self.root, data)

def _search(self, node, data):
    if node is None:
        return f"The Search node is not found!!!"
    
    if node.data == data:
        return f"the Search node {data} is Found"
    
    if data < node.data:
        return self._search(node.left, data)
    elif data > node.data:
        return self._search(node.right, data)

def search_iterative(self, data): # Iterative
    current = self.root
     
    while current is not None:
        if current.data == data:
            return True
        
        if data < current.data:
            current = current.left
        else:
            current = current.right

    return False

    
BST.search = search
BST._search = _search
BST.search_iterative = search_iterative
print(tree.search(10))
print(tree.search(100))
print(tree.search(90))
print(tree.search_iterative(10))
print(tree.search_iterative(101))
```

```python
""" 
c. Delete in BST (3 Cases)
    -Case 1 — Delete a Leaf Node
    -Case 2 — Delete a Node with One Child
    -Case 3 — Delete a Node with Two Children


Time Complexity
Case	       Time        Complexit Why
Best Case	    O(1)	    The node to delete is the root and has 0 or 1 child.
Average Case	O(log n)	In a balanced BST, we travel from the root to the node (tree height = log n).
Worst Case	    O(n)	    In a skewed BST (like a linked list), we may traverse all n nodes.

Space Complexity
Case	Space Complexity
Best	O(1)
Average	O(log n)
Worst	O(n)
"""
def delete(self, data):
    self.root = self._delete(self.root, data)

def _delete(self, node, data):

    if node is None:
        return None

    if data < node.data:
        node.left = self._delete(node.left, data)

    elif data > node.data:
        node.right = self._delete(node.right, data)

    else:

        if node.left is None:
            return node.right

        elif node.right is None:
            return node.left

        successor = self._find_min(node.right)
        node.data = successor.data
        node.right = self._delete(node.right, successor.data)

    return node

def _find_min(self, node):
    while node.left:
        node = node.left
    return node

BST.delete = delete
BST._delete = _delete
BST._find_min = _find_min

tree.delete(95)
tree.inorder(tree.root)
print()
tree.delete(50)
tree.inorder(tree.root)


```

```python
""" 
Find Closest Value in BST

    1. GPS / Nearest Location
    2. Student Marks
    3. Product Price Search

Time Complexity
Case	Complexity
Best	O(1)
Average	O(log n)
Worst	O(n)

Space:
| Recursive | O(log n) average, O(n) worst |
| Iterative | O(1) ✅ |
"""
def find_closest_value(self, target):
   current = self.root 
   closest = self.root.data

   while current:
        if abs(target - current.data) < abs(target - closest):
            closest = current.data
        if target < current.data:
            current = current.left
        elif target > current.data:
            current = current.right
        
        else:
            break
   return closest

BST.find_closest_value = find_closest_value
print(tree.inorder(tree.root))
print(tree.root.data, "- root")
print()
print(tree.find_closest_value(59))
```

```python
""" 
Validate BST (is_valid_bst())
    "Given a binary tree, determine whether it is a valid Binary Search Tree."

_is_valid_bst(50, -∞, +∞)
│
├── _is_valid_bst(30, -∞, 50)
│   │
│   ├── _is_valid_bst(20, -∞, 30)
│   │   ├── _is_valid_bst(None, -∞, 20) → True
│   │   └── _is_valid_bst(None, 20, 30) → True
│   │
│   └── _is_valid_bst(40, 30, 50)
│       ├── _is_valid_bst(None, 30, 40) → True
│       └── _is_valid_bst(None, 40, 50) → True
│
└── _is_valid_bst(70, 50, +∞)
    │
    ├── _is_valid_bst(60, 50, 70)
    │   ├── _is_valid_bst(None, 50, 60) → True
    │   └── _is_valid_bst(None, 60, 70) → True
    │
    └── _is_valid_bst(80, 70, +∞)
        ├── _is_valid_bst(None, 70, 80) → True
        └── _is_valid_bst(None, 80, +∞) → True
Time Complexity
Case	Complexity
Best	O(n)
Average	O(n)
Worst	O(n)

Space Complexity
Case	        Complexity
Balanced BST	O(log n)
Skewed BST	    O(n)
"""
def is_valid_bst(self):
    return self._is_valid_bst(self.root, float("-inf"), float("inf"))

def _is_valid_bst(self, node, minimum, maximum):
    if node is None:
        return True

    if not (minimum < node.data < maximum):
        return False 
    
    return (
        self._is_valid_bst(node.left, minimum, node.data)
        and
        self._is_valid_bst(node.right, node.data, maximum)
    )

BST.is_valid_bst = is_valid_bst
BST._is_valid_bst = _is_valid_bst
print(tree.is_valid_bst())
```

```python
""" 
kth_smallest()
    - This is one of the most common BST interview questions.


Time Complexity
Case	Complexity
Best	O(n)
Average	O(n)
Worst	O(n)

Every node is visited exactly once.

Space Complexity
Case	      Complexity
Result List 	-    O(n)
Recursion Stack - 	 O(log n) average, O(n) worst
"""
def Kth_smallest(self, k):
    result = []
    self._inorder(self.root, result)
    return result[k-1]

def _inorder(self, node, result):
    if node is None:
        return 
    
    self._inorder(node.left, result)
    result.append(node.data)
    self._inorder(node.right, result)

BST.Kth_smallest = Kth_smallest
BST._inorder = _inorder
print(tree.Kth_smallest(2))
```
