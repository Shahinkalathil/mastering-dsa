# Bubble Sort syntax in Python
1. Student Marks Sorting
2. Sensor Readings (Low Memory System)

```python
# Bubble Sort syntax in Python
def Bubble_sort(arr):
	n = len(arr)

	for i in range(n):
		for j in range(0, n-i-1):
			if arr[j] > arr[j+1]:
				arr[j], arr[j+1] = arr[j+1], arr[j]
	return arr


arr = [5,3,4,1,2]
print(f"The arry is {arr}.")
print(f"The array after Bubble sort {Bubble_sort(arr)}.")
```

```python
# 1. Student Marks Sorting
def bubble_sort_marks(marks):
    n = len(marks)
    
    for i in range(n):
        for j in range(0, n-i-1):
            if marks[j] > marks[i]:
                marks[j], marks[j+1] = marks[j+1], marks[i]
    return marks

# Input
marks = [78, 45, 90, 32, 67]

# Process
sorted_marks = bubble_sort_marks(marks)

# Output
print("Unsorted Marks:", marks)
print("Sorted Marks:", sorted_marks)
```

```python
# 2. Sensor Readings (Low Memory System)
def bubble_sort_sensor(data):
    n = len(data)
    
    for i in range(n):
        for j in range(0, n-i-1):
            if data[j] > data[j+1]:
                temp = data[j]
                data[j] = data[j+1]
                data[j+1] = temp
    return data

# Input
sensor_data = [55, 23, 87, 12, 34]

# Process
sorted_sensor = bubble_sort_sensor(sensor_data)

# Output
print("Unsorted Sensor Data", sensor_data)
print("Sorted Sensor Data:", sorted_sensor)
```
