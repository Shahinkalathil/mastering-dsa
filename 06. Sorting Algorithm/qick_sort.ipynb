# Quick Sort in Python 👇

```python
# 🔁 Recursive Formula
# quick sort(A) = quick sort(A<pivot) + pivot + quick sort(A>pivot)

def partition(arr, st, end):
    idx = st - 1
    pivot = arr[end]

    for j in range(st, end):
        if arr[j] <= pivot:
            idx += 1
            arr[idx],  arr[j] = arr[j], arr[idx]
    arr[idx+1], arr[end] = arr[end], arr[idx+1]
    return idx + 1

def quick_sort(arr, st, end):
    if st < end:

        pivIdx = partition(arr, st, end)
        quick_sort(arr, st, pivIdx-1)
        quick_sort(arr, pivIdx + 1, end)

arr = [5, 2, 6, 4, 1, 3]
quick_sort(arr, 0, len(arr)-1)
print(arr)
```
