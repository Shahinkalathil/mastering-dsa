# Concept of Trie Data Structure
1. Trie Insert & Search
2. Prefix Search (startsWith)
3. Count Words with Given Prefix

```python
""" 
1. Trie Insert & Search
"""
class Trie_Node:
    def __init__(self):
        self.children = {}
        self.is_end = False

class Trie:
    def __init__(self):
        self.root = Trie_Node()


    def insert(self, word):
        current = self.root

        for char in word:
            if char not in current.children:
                current.children[char] = Trie_Node()

            current = current.children[char]
        current.is_end = True

    def search(self, word):
        current = self.root

        for char in word:
            if char not in current.children:
                return False
            current = current.children[char]
        return current.is_end

words = [
    "apple",
    "app",
    "apply",
    "ape",
    "ear",
    "earn",
    "earth",
    "rat",
    "rate",
    "rating",
    "telephone",
    "phone",
    "phoneshop",
    "phonecall"
]

trie = Trie()
for word in words:
    trie.insert(word)

print(trie.search("shahin"))
print(trie.search("apple"))

print(trie.root.children)

```

```python
def start_with(self, prefix):
    current = self.root

    for char in prefix:
        if char not in current.children:
            return False
        current = current.children[char]

    return True

Trie.start_with = start_with
print(trie.start_with("apple"))
print(trie.start_with("ple"))
```

```python
def autocomplete(self, prefix):
    current = self.root

    for char in prefix:
        if char not in current.children:
            return []
        current = current.children[char]

    words = []
    self._dfs(current, prefix, words)
    return words

def _dfs(self, node, word, words):
    if node.is_end:
        words.append(word)
    for char, child in node.children.items():
        self._dfs(child, word + char, words)

Trie.autocomplete = autocomplete
Trie._dfs = _dfs

print(trie.autocomplete("app"))

```
