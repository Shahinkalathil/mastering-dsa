# Concepts of MinHeap
1. Implement a Min Heap from scratch (array-based).
2. Implement insert() in a Min Heap using heapify-up (sift-up).
3. Implement remove_min() in a Min Heap using heapify-down (sift-down).


```python
class MinHeap:
    def __init__(self):
        self.heap = []

    def _parent(self, index):
        return (index - 1) // 2
    
    def _left_child(self, index):
        return 2 * index + 1
     
    def _right_child(self, index):
        return 2 * index + 2
    
    def _swap(self, i, j):
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]
```

```python
""" 
Heapify Up Algorithm
    Suppose the inserted element is at index 5.
    If the heap property is already satisfied, stop immediately.
    No more swaps are needed.   
Time Complexity
Case	Complexity
Best	O(1)
Average	O(log n)
Worst	O(log n)
"""
def _heapify_up(self, index):
    while index > 0:
        parent = self._parent(index)

        if self.heap[index] < self.heap[parent]:
            self._swap(index, parent)
            index = parent
        else:
            break
MinHeap._heapify_up = _heapify_up
```

```python
""" 
insert()

Time Complexity
Case	Complexity
Best	O(1)
Average	O(log n)
Worst	O(log n)
"""
def insert(self, value):
    self.heap.append(value)
    self._heapify_up(len(self.heap)-1)

MinHeap.insert  = insert
heap_value = MinHeap()

values = [10, 20, 30, 40, 50, 5]
for val in values:
    heap_value.insert(val)

print(heap_value.heap)
```

```python
""" 
Heapify Down Algorithm 
    Heapify Down is the process of moving an element downward until the Min Heap property is restored.
    It is used after removing the root.

Time Complexity
Case	Complexity
Best	O(1)
Average	O(log n)
Worst	O(log n)
"""
def _heapify_down(self, index=0):
    size = len(self.heap) 

    while True:
        smallest = index
        left = self._left_child(index)
        right = self._right_child(index)

        if left < size and self.heap[left] < self.heap[smallest]:
            smallest = left

        if right < size and self.heap[right] < self.heap[smallest]:
            smallest = right

        if smallest != index:
            self._swap(index, smallest)
            index = smallest
        else:
            break

MinHeap._heapify_down = _heapify_down
```

```python
""" 
remove_min()
    remove_min() removes and returns the smallest element from the heap.
    In a Min Heap, the smallest element is always at the root (index 0).

Time Complexity
Case	Complexity
Best	O(1)
Average	O(log n)
Worst	O(log n)
"""

def remove_min(self):
    if not self.heap:
        return None
    
    if len(self.heap) == 1:
        self.heap.pop()

    min_value = self.heap[0]
    self.heap[0] = self.heap.pop()
    self._heapify_down()
    return f"this is the value we are removed from the array {min_value}"

MinHeap.remove_min = remove_min
print(heap_value.heap)
heap_value.remove_min()
print()
print(heap_value.heap)
```

```python
def peek(self):
    if self.heap is None:
        return None
    return self.heap[0]

MinHeap.peek = peek
print(heap_value.peek())
```
