# Concepts of Heap
1. Min Heap
2. Max Heap
3. Build Heap from List

```python
# Min Heap
import heapq

# Create Min Heap
min_heap = []

# Insert
heapq.heappush(min_heap, 10)
heapq.heappush(min_heap, 5)
heapq.heappush(min_heap, 20)

print("Min Heap:", min_heap)

# Remove (smallest)
print("Removed:", heapq.heappop(min_heap))
print("After Removal:", min_heap)
```

```python
# Max Heap
import heapq

# Max Heap using negative values
max_heap = []

heapq.heappush(max_heap, -10)
heapq.heappush(max_heap, -5)
heapq.heappush(max_heap, -20)

print("Max Heap:", [-x for x in max_heap])

# Remove (largest)
print("Removed:", -heapq.heappop(max_heap))
print("After Removal:", [-x for x in max_heap])
```

```python
# Build Heap from List
import heapq

arr = [40, 10, 30, 20]

heapq.heapify(arr)

print("Heapified Array:", arr)
```
