```python
class MaxHeap():
    def __init__(self):
        self.heap = []

    def _left_child(self, index):
        return 2 * index + 1 
    def _right_child(self, index):
        return 2 * index + 2
    def _swap(self, i, j):
         self.heap[i], self.heap[j] = self.heap[j], self.heap[i]

    def _heapify_down(self, index, size):
        
        while True:
            largest = index
            left = self._left_child(index)
            right = self._right_child(index)

            if left < size and self.heap[left] > self.heap[largest]:
                largest = left
            if right < size and self.heap[right] > self.heap[largest]:
                largest = right 

            if largest != index:
                self._swap(index, largest)
                index = largest
            else:
                break

    def build_heap(self):
        n = len(self.heap)
        for i in range((n//2)-1, -1, -1):
            self._heapify_down(i, n)
        
    def heap_sort(self):
        self.build_heap()

        for end in range(len(self.heap)-1, 0, -1):
            self._swap(0, end)

            self._heapify_down(0, end)

        return self.heap



heap = MaxHeap()

heap.heap = [5, 3, 8, 1, 4, 9, 2]

print("Original :", heap.heap)

sorted_array = heap.heap_sort()

print("Sorted in Asending order :", sorted_array) 

```

```python
""" 
Q 1. Sort an array of strings alphabetically using Heap Sort.
"""
class MaxHeap:
    def __init__(self):
        self.heap = []

    def _left_child(self, index):
        return 2 * index + 1
    
    def _right_child(self, index):
        return 2 * index + 2 
    
    def _swap(self, i, j):
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]

    def _heapify_down(self, index, size):

        while True:
            largest = index
            left = self._left_child(index)
            right = self._right_child(index)

            if left < size and self.heap[left] > self.heap[largest]:
                largest = left

            if right < size and self.heap[right] > self.heap[largest]:
                largest = right

            if largest != index:
                self._swap(index, largest)
                index = largest

            else:
                break

    def build_heap(self):
        n = len(self.heap)
        for i in range((n//2)-1, -1, -1):
            self._heapify_down(i, n)


    def heap_sort(self):
        self.build_heap()
        
        for end in range(len(self.heap)-1, 0, -1):
            self._swap(0, end)

            self._heapify_down(0, end)

        return self.heap
    
heap = MaxHeap()

heap.heap = ["orange", "apple", "banana", "grape", "kiwi"]

print(heap.heap_sort())
```
