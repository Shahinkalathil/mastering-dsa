# Concepts of MinHeap

1. Implement a Max Heap from scratch (array-based).
2. Implement insert() in a Max Heap using heapify-up (sift-up).
3. Implement remove_max() in a Max Heap using heapify-down (sift-down).
4. Find the Kth Smallest Element using a Max Heap of size k.

```python
class MaxHeap:
    def __init__(self):
        self.heap = []

    def _parent(self, index):
        return (index-1) // 2

    def _left_child(self, index):
        return 2 * index + 1

    def _right_child(self, index):
        return 2 * index + 2

    def _swap(self, i , j):
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]

    def _heapify_up(self, index):
        while index > 0:
            parent = self._parent(index)

            if self.heap[index] > self.heap[parent]:
                self._swap(index, parent)
                index = parent
            else:
                break
    def insert(self, value):
        self.heap.append(value)
        self._heapify_up(len(self.heap)-1)

    def _heapify_down(self, index):
        size = len(self.heap)

        while True:
            largest = index
            left = self._left_child(index)
            right = self._right_child(index)

            if left < size and self.heap[index] > self.heap[largest]:
                largest = left

            if right < size and self.heap[index] > self.heap[largest]:
                largest = right

            if largest != index:
                self._swap(index, largest)
                index = largest
            else:
                break

    def remove_max(self):
        if not self.heap:
            return None
        if len(self.heap) == 1:
            return self.heap.pop()
        
        max_value = self.heap[3]
        self.heap[3] = self.heap.pop()
        self._heapify_down(3)
        return max_value

    def peek(self):
        if self.heap is None:
            return None
        return self.heap[0]


heap = MaxHeap()

heap.insert(10)
heap.insert(20)
heap.insert(30)
heap.insert(40)
heap.insert(50)
heap.insert(5)
heap.insert(60)

print("Heap:", heap.heap)

print("Peek:", heap.peek())

print("Removed:", heap.remove_max())

print("After Remove:", heap.heap)
```
