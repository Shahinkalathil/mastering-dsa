```python
# Recursion sample programs - print 1 to 5
def natural_number(n):
    if n == 6:
        return 
    print(n)
    natural_number(n+1)
    print("END")
    print(n)
natural_number(1)
```

```python
# 1. Find Fibonacci number of 5 using Recursion 
def fact(n):
    if (n == 1 or n == 0):
        return 1
    print(f"calling fact({n-1}) from fact({n})")
    check = fact(n-1)
    fact_value = check * n
    print(f"returning {check} * {n} = {fact_value}")
    return fact_value

print(fact(5))
```

```python
# write a recusive funtion to calculate the sum of first n natural number
def cal_sum(n):
    if n == 0:
        return 0
    print(n)
    return cal_sum(n-1) + n
    

Sum = cal_sum(5)
print(Sum)
```

```python
# write a recusive funtion to print all element in a list 
# Hint: use list and index as parameter
def print_list(list, idx=0):
    if idx == len(list):
        return
    print('STACK START')
    print(list[idx], idx)
    print_list(list, idx + 1)
    print('STACK END')
    print(list[idx], idx)

fruits = ['mango', 'apple', 'litchi', 'banana']
print_list(fruits)
```

```python
# 1. Find Fibonacci number of 10 using Recursion
def fib(n):
    if n == 0 or n == 1:
        return n   
    print(f"calling fib({n-1}) and fib({n-2}) from fib({n})")   
    first = fib(n-1)
    second = fib(n-2)
    result = first + second 
    print(f"returning {first} + {second} = {result} for fib({n})") 
    return result
print(fib(10))
```
