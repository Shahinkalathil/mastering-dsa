```python
# Recursion sample programs - print 1 to 5
def natural_number(n):
    if n == 6:
        return 
    print(n)
    natural_number(n+1)
    print("END")
natural_number(1)
```

```python
# 1. Find Fibonacci number of 5 using Recursion 
def fact(n):
    if (n==1 or n==0):
        return 1
    return fact(n-1) * n
    

print(fact(5))
```

```python
# write a recusive funtion to calculate the sum of first n natural number
def cal_sum(n):
    if n == 0:
        return 0
    print(n)
    return cal_sum(n-1) + n
    

Sum = cal_sum(5)
print(Sum)
```

```python
# write a recusive funtion to print all element in a list 
# Hint: use list and index as parameter
def print_list(list, idx=0):
    if idx == len(list):
        return
    print('STACK START')
    print(list[idx], idx)
    print_list(list, idx + 1)
    print('STACK END')
    print(list[idx], idx)

fruits = ['mango', 'apple', 'litchi', 'banana']
print_list(fruits)
```
