```python
# 1. Find factorial number of 5 using Recursion 
def fact(n):
    if (n == 1 or n == 0):
        return 1
    print(f"calling fact({n-1}) from fact({n})")
    check = fact(n-1)
    fact_value = check * n
    print(f"returning {check} * {n} = {fact_value}")
    return fact_value

print(fact(5))
```

```python
# 2. write a recusive funtion to calculate the sum of first n natural number
def cal_sum(n):
    if n == 0:
        return 0
    print(n)
    return cal_sum(n-1) + n
    

Sum = cal_sum(5)
print(Sum)
```

```python
# write a recusive funtion to print all element in a list 
# Hint: use list and index as parameter
def print_list(list, idx=0):
    if idx == len(list):
        return
    print('STACK START')
    print(list[idx], idx)
    print_list(list, idx + 1)
    print('STACK END')
    print(list[idx], idx)

fruits = ['mango', 'apple', 'litchi', 'banana']
print_list(fruits)
```

```python
# 1. Find Fibonacci number of 10 using Recursion
def fib(n):
    if n == 0 or n == 1:
        return n    
    print("first func  :  ", n) 
    first = fib(n-1)
    print("second func  :  ", first)
    second = fib(n-2)
    print(3, "  :  ", first,"+" , second )
    result = first + second 
    print(4, "  :  ", result)
    print(f"returning {first} + {second} = {result} for fib({n})") 
    print(5)
    return result
print(fib(5))
```

```python
# 2. Find sum of this array arr = [2,4,6,3,1,44,32] using Recursion
def cal_sum(arr, idx=0):
    if idx == len(arr):
        return 0
    return cal_sum(arr, idx+1) + arr[idx]
    

arr = [1,4,2,5,6,7]
print(f"the sum of the arr is {cal_sum(arr)}")
```

```python
# 3. Find a number is prime using Recursion
def prime_finder(n, div=None):
    if n <= 1:
        return True
    if div is None:
        div = int(n ** 0.5)
    if div == 1:
        return True
    if n % div == 0:
        return False
    return prime_finder(n, div-1)

n = 3
if prime_finder(n):
    print(f"{n} is a prime number")
else: 
    print(f"{n} is not a prime number")
```
